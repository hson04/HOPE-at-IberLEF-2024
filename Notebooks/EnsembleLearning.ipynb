```python
df1 = pd.read_csv("/kaggle/input/ensemble-learning-task-2b-spanish/1.csv")
df2 = pd.read_csv("/kaggle/input/ensemble-learning-task-2b-spanish/2.csv")
df3 = pd.read_csv("/kaggle/input/ensemble-learning-task-2b-spanish/3.csv")
df4 = pd.read_csv("/kaggle/input/ensemble-learning-task-2b-spanish/4.csv")
df5 = pd.read_csv("/kaggle/input/ensemble-learning-task-2b-spanish/5.csv")
df6 = pd.read_csv("/kaggle/input/ensemble-learning-task-2b-spanish/6.csv")
df7 = pd.read_csv("/kaggle/input/ensemble-learning-task-2b-spanish/7.csv")
```

```python
df1.rename(columns={'category': 'category 1'}, inplace=True)
df2.rename(columns={'category': 'category 2'}, inplace=True)
df3.rename(columns={'category': 'category 3'}, inplace=True)
df4.rename(columns={'category': 'category 4'}, inplace=True)
df5.rename(columns={'category': 'category 5'}, inplace=True)
df6.rename(columns={'category': 'category 6'}, inplace=True)
df7.rename(columns={'category': 'category 7'}, inplace=True)
```

```python
df1
```

```python
df1 = df1.merge(df2[['id', 'category 2']], how='left', on='id')
df1 = df1.merge(df3[['id', 'category 3']], how='left', on='id')
df1 = df1.merge(df4[['id', 'category 4']], how='left', on='id')
df1 = df1.merge(df5[['id', 'category 5']], how='left', on='id')
df1 = df1.merge(df6[['id', 'category 6']], how='left', on='id')
df1 = df1.merge(df7[['id', 'category 7']], how='left', on='id')
```

```python

def ensemble_vote(row):
    predictions = row[1:]
    prediction_counts = predictions.value_counts().to_dict()
    winning_category = max(prediction_counts, key=prediction_counts.get)

    return winning_category


df1['category'] = df1.apply(ensemble_vote, axis=1)
```

```python
df1
```

```python
df = df1
columns_to_keep = ['id', 'category']  # List of columns to retain (delete the rest)

df.drop(columns=[col for col in df.columns if col not in columns_to_keep], axis=1, inplace=True)
```
